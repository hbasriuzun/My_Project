```python
import pandas as pd
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif  # Özelliklerin önemini değerlendirmek için kullanılan bir istatistik test

# Veri kümesini yükleyin (örnek olarak bir CSV dosyası kullanıyoruz)
veri = pd.read_csv('ayrilan_veri.csv')

# Bağımsız değişkenleri ve hedef değişkeni ayırın
X = veri.drop('Hedef_Degisken', axis=1)  # Bağımsız değişkenler
y = veri['Hedef_Degisken']  # Hedef değişken

# SelectKBest ile en iyi k özellik seçimi yapın (k'yi istediğiniz sayıya ayarlayın)
k = 5
seçici = SelectKBest(score_func=f_classif, k=k)
X_new = seçici.fit_transform(X, y)

# En iyi k özelliğin adını alın
en_iyi_ozellikler = X.columns[seçici.get_support(indices=True)]

# Seçilen en iyi k özellikleri yazdırın
print(f"Seçilen En İyi {k} Özellikler: {en_iyi_ozellikler.tolist()}")
```

```python
import time

def recursive_fib(n):
    if n == 0 or n == 1:
        return n
    return recursive_fib(n - 1) + recursive_fib(n - 2)

def fib_On(n):
    a, b = 0, 1
    for i in range(2, n + 1):
        c = a + b
        a = b
        b = c
    return c

def fib_Ologn(n):
    def multiply(F, M):
        x = F[0][0] * M[0][0] + F[0][1] * M[1][0]
        y = F[0][0] * M[0][1] + F[0][1] * M[1][1]
        z = F[1][0] * M[0][0] + F[1][1] * M[1][0]
        w = F[1][0] * M[0][1] + F[1][1] * M[1][1]

        F[0][0] = x
        F[0][1] = y
        F[1][0] = z
        F[1][1] = w

    def power(F, n):
        M = [[1, 1], [1, 0]]
        if n == 0 or n == 1:
            return
        power(F, n // 2)
        multiply(F, F)
        if n % 2 != 0:
            multiply(F, M)

    F = [[1, 1], [1, 0]]
    if n == 0:
        return 0
    power(F, n - 1)
    return F[0][0]

baslangic_zaman = time.time()
result = fib_On(20000)
bitis_zaman = time.time()

print("Sonuç:", result)
print("İşlem süresi:", (bitis_zaman - baslangic_zaman) * 1000000, "mikrosaniye")
```

```python
baslangic_zaman = time.time()
result = fib_Ologn(20000)
bitis_zaman = time.time()

print("Sonuç:", result)
print("İşlem süresi:", (bitis_zaman - baslangic_zaman) * 1000000, "mikrosaniye")
```
